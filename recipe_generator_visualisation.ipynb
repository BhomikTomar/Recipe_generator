```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/Recipe_Dataset_Final.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the relevant columns for visualization
columns = ['calories', 'fat', 'protein', 'sodium']
data = df[columns]

# Set up the figure size
plt.figure(figsize=(15, 5))

# Heatmap for correlation analysis
plt.subplot(1, 3, 1)
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")

# Boxplot for each feature to detect outliers
plt.subplot(1, 3, 2)
sns.boxplot(data=data)
plt.title("Boxplot of Nutritional Features")
plt.xticks(rotation=45)

# Histograms for each feature
plt.subplot(1, 3, 3)
data.hist(bins=20, figsize=(15, 5), layout=(1, 4))
plt.suptitle("Histograms of Nutritional Features")

# Show the plots
plt.tight_layout()
plt.show()
```
